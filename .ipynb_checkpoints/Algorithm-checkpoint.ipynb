Simple Linear

```python
import pandas as pd
df=pd.read_csv("placement.csv")
df
```

```python
df.isnull().sum()
```

```python
#pip install scikit-learn
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["cgpa"],df["package"])
plt.xlabel("cgpa")
plt.ylabel("package")
```

```python
x=df[['cgpa']]#independant feature/ i\p
y=df[['package']]#dependant feature/ o\p
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=39)
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
print(dir(model))
```

```python
model.fit(x_train,y_train)
```

```python
#train acc
model.score(x_train,y_train)
```

```python
model.predict([[6.89]]),model.predict([[5.66]])
```

```python
model.predict([[5.66]])
```

```python
#y_test
```

```python
# this is my bulk prediction
y_pred_test=model.predict(x_test)
y_pred_test,y_test
```

```python
from sklearn.metrics import r2_score
```

```python
#test acc
r2_score(y_test,y_pred_test)
```

```python
model.predict([[9.0]])
```
